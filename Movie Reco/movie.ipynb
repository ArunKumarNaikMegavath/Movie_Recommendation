```python
import numpy as np
import pandas as pd
import diffliby
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie=pd.read_csv("movies.csv")
movie.head()
```

```python
movie.shape
```

Selecting features

```python
selected_featurs=['genres','keywords' , 'tagline','cast', 'director']
selected_featurs
```

replacing null values in features

```python
for need in selected_featurs:
    movie[need]=movie[need].fillna("")
```

combining all selected features

```python
comb_fearure=movie['genres']+movie['keywords']+movie['tagline']+movie['cast']+movie['director']
```

```python
comb_fearure
```

Text to Vecrtor

```python
Vectorizer=TfidfVectorizer
```

```python
feature_vec = TfidfVectorizer().fit_transform(comb_fearure)
print(feature_vec)
```

similarty

```python
similarty=cosine_similarity(feature_vec)
```

```python
print(similarty)
```

```python
print(similarty.shape)
```

User movie Name

```python
movie_name=input("Enter you Favorite Movie Name:")
print(movie_name)
```

```python
list_of_movie=movie["title"].tolist()
print(list_of_movie)
```

find closest match  

```python
list_of_closest_match=difflib.get_close_matches(movie_name,list_of_movie)
print(list_of_closest_match)
```

```python
close_match=list_of_closest_match[0]
print(close_match)
```

find the index of the movie

```python
index_of_movie=movie[movie.title==close_match]['index'].values[0]
print(index_of_movie)
```

SIMILARITY Score

```python
similarity_score=list(enumerate(similarty[index_of_movie]))
print(similarity_score)
```

sort


```python
sorted_sim_movie=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_sim_movie)
```

```python
print("Movie Suggestions for you :\n")

i=1
for movies in sorted_sim_movie:
    index=movies[0]
    title_from_index=movie[movie.index==index]['title'].values[0]
    if (i<=30):
        print(i,'.',title_from_index)
        i+=1
```

MOVIE RECOMMENDATION SYSTEM

```python
movie_name=input("Enter the Movie Name:")
list_of_movie=movie['title'].tolist()
list_of_closest_match=difflib.get_close_matches(movie_name,list_of_movie)
close_match=list_of_closest_match[0]
index_of_movie=movie[movie.title==close_match]['index'].values[0]
similarity_score=list(enumerate(similarty[index_of_movie]))
sorted_sim_movie=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print("Movie Suggestions for you :\n")

i=1
for movies in sorted_sim_movie:
    index=movies[0]
    title_from_index=movie[movie.index==index]['title'].values[0]
    if (i<=30):
        print(i,'.',title_from_index)
        i+=1

```
